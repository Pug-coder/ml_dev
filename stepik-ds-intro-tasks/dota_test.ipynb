```python
df = pd.read_csv('http://stepik.org/media/attachments/course/4852/algae.csv')
df.head()
```

```python
concentrations = df.groupby('group',as_index=True)
concentrations.agg({
    'sucrose': lambda x: max(x) - min(x),
    'group': 'count', 
    'citrate': 'var'
    })
```

```python
mean_concentrations = concentrations.agg({
    'sucrose' : 'mean',
    'alanin' : 'mean',
    'citrate' : 'mean',
    'glucose' : 'mean',
    'oleic_acid' : 'mean'
})
mean_concentrations.head()
```
